IMPORT LIBRAIRIES AND VIDEO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy.signal import find_peaks

im = io.imread('sortie_23k.tif')
im=255-im
```

DISPLAY AN IMAGE

```python
plt.imshow(im[1000,:,:], cmap='gray')
```

TAKE A VERTICAL LINE OF THE VIDEO AND PLOT THE TEMPORAL TRACE

```python
cut=im[:,50:80,300]
cut2=np.sum(cut,1)
plt.plot(cut2[100:200])
```

FIND THE PEAKS ON A SMALL TRACE

```python
x=cut2[1:100]
peaks2, _ = find_peaks(x, height=3000)   
plt.plot(peaks2, x[peaks2], "ob"); plt.plot(x); plt.legend(['prominence'])
print(peaks2.size)
```

APPLY IT TO THE ENTIRE VIDEO

```python
x=cut2
peaks2, _ = find_peaks(x, height=3000) 
plt.plot(peaks2, x[peaks2], "ob"); plt.plot(x); plt.legend(['prominence'])
```

CALCULATE THE FLOW RATE

```python
fps=23000
print('nombre de capsules =',peaks2.size)
temps_total=im.shape[0]/fps
print('temps total =',temps_total)
debit=peaks2.size/temps_total
print('nombre de capsules /s=',debit)
```
